```python

!pip install gtts gradio deep-translator --quiet

import gradio as gr
from gtts import gTTS
from deep_translator import GoogleTranslator
import time

# Popular countries' languages (gTTS supported)
POPULAR_LANGUAGES = {
    "English (US/UK/Australia)": "en",
    "Japanese (Japan)": "ja",
    "Thai (Thailand)": "th",
    "Myanmar (Myanmar)": "my",
    "French (France)": "fr",
    "German (Germany)": "de",
    "Spanish (Spain/Latin America)": "es",
    "Chinese (Mandarin)": "zh-cn",
    "Korean (South Korea)": "ko",
    "Russian (Russia)": "ru",
    "Italian (Italy)": "it",
    "Arabic (Saudi/Egypt/UAE)": "ar",
    "Hindi (India)": "hi",
    "Portuguese (Portugal/Brazil)": "pt",
    "Vietnamese (Vietnam)": "vi"
}

def tts_mixed_text(text, target_language):
    if not text.strip():
        return None, ""

    # Detect if text contains Myanmar or English, translate everything
    translated_text = GoogleTranslator(source='auto', target=POPULAR_LANGUAGES[target_language]).translate(text)

    # Generate TTS audio
    filename = f"voice_{int(time.time())}.mp3"
    tts = gTTS(text=translated_text, lang=POPULAR_LANGUAGES[target_language])
    tts.save(filename)

    return filename, translated_text

# Gradio interface
app = gr.Interface(
    fn=tts_mixed_text,
    inputs=[
        gr.Textbox(
            label="မြန်မာ + English Text ရိုက်ထည့်ပါ",
            lines=5,
            placeholder="ဥပမာ - ကျောင်း and school"
        ),
        gr.Dropdown(
            choices=list(POPULAR_LANGUAGES.keys()),
            label="အသံဘာသာစကား (Popular Countries)",
            value="English (US/UK/Australia)"
        )
    ],
    outputs=[
        gr.Audio(
            label="အသံထွက် (Play / Download)",
            type="filepath"
        ),
        gr.Textbox(
            label="ဘာသာပြန်စာသား (Translated Text)",
            lines=2,
            interactive=False
        )
    ],
    title="Mixed Myanmar + English Text-to-Speech",
    description="TextBox မှာ မြန်မာ + English text → Popular Language ရွေး → Audio + Translated Text"
)

app.launch(share=True, inbrowser=True)
```
